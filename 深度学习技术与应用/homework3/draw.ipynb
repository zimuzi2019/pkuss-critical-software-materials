```python
import nbimporter

from CRNN import load_model, load_weights, CRNN
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = load_model('0123456789', device, seq_proj=[7, 30], backend='resnet18', snapshot='crnn_resnet18_0123456789_best').to(device)
```

```python
from torchsummary import summary


summary(model, input_size=(3, 100, 200)) 
```

```python
import torch.onnx
import netron

torch.onnx.export(model, torch.zeros(3, 3, 100, 200).to(device), "CRNN.onnx", verbose=True)

netron.start('CRNN.onnx')
```
